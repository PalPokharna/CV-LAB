```python
import numpy as np
import skimage
from skimage.io import imread
from skimage import io
from skimage.feature import greycomatrix
from skimage.color import rgb2gray
import imageio
import cv2
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread('smooth1.jpg')
io.imshow(image)
io.show()
```

```python
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
io.imshow(image)
io.show()
```

```python
image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
io.imshow(image)
io.show()
```

```python
co_matrix = skimage.feature.greycomatrix(image, [5], [0], levels=256, symmetric=True, normed=True)
# Calculate texture features from the co-occurrence matrix
contrast = skimage.feature.greycoprops(co_matrix, 'contrast')
correlation = skimage.feature.greycoprops(co_matrix, 'correlation')
energy = skimage.feature.greycoprops(co_matrix, 'energy')
homogeneity = skimage.feature.greycoprops(co_matrix, 'homogeneity')
# Print the texture features
print("Contrast:", contrast)
print("Correlation:", correlation)
print("Energy:", energy)
print("Homogeneity:",homogeneity)
```

```python
import numpy as np
from patchify import patchify
from PIL import Image

image = np.asarray(image2)
height, width = image.shape[:2]

patch_height = height // 2
patch_width = width // 3
patches = patchify(image, (patch_height, patch_width), step=patch_width)

# Combine the patches into a single image
combined_image = np.vstack([np.hstack(row) for row in patches])

# Convert the combined image back to PIL Image
combined_image_pil = Image.fromarray(combined_image)

# Display the combined image
combined_image_pil.show()
```
